```python
def parse_mrz(mrz: list):
    # check the length of the first line to determine the type of MRZ
    if len(mrz[0]) == 44:
        # Type 1 MRZ
        mrz_type = 1
        pattern = r'([A-Z])([A-Z<])([A-Z]{3})([A-Z<]+)<<([A-Z<]+)(\d{6})(\d)([A-Z]{3})(\d{6})(\d)([A-Z])(\d{6})(\d)(\w+)'
    elif len(mrz[0]) == 36:
        # Type 2 MRZ
        mrz_type = 2
        pattern = r'([A-Z])([A-Z<])([A-Z]{3})([A-Z<]+)<<([A-Z<]+)(\d{6})(\d)([A-Z]{3})(\d{6})(\d)([A-Z])(\d{6})(\d)'
    elif len(mrz[0]) == 30:
        # Type 3 MRZ
        mrz_type = 3
        pattern = r'([A-Z])([A-Z<])([A-Z]{3})([A-Z<]+)<<([A-Z<]+)(\d{9})(\d)([A-Z]{3})(\d{6})(\d)([A-Z])(\d{6})(\d)(\w+)'
    else:
        # unknown MRZ type
        return None
    
    # join the lines of the MRZ into a single string
    mrz = ''.join(mrz)
    
    # try to match the MRZ with the determined type
    match = re.match(pattern, mrz)
    
    # check if a match was found
    if not match:
        # if no match was found, return None
        return None
    
    # parse the information from the MRZ
    groups = match.groups()
    passport_type = groups[0]
    passport_subtype = groups[1]
    issuing_country = groups[2]
    surname = groups[3].replace('<', ' ').split()
    given_names = groups[4].replace('<', ' ').split()
    passport_number = groups[5]
    passport_number_check_digit = groups[6]
    nationality = groups[7]
    date_of_birth = groups[8]
    date_of_birth_check_digit = groups[9]
    sex = groups[10]
    expiration_date = groups[11]
    expiration_date_check_digit = groups[12]
    
    if mrz_type == 1:
        personal_number = groups[13]
        personal_number_check_digit = None
    elif mrz_type == 2:
        personal_number = None
        personal_number_check_digit = None
    elif mrz_type == 3:
        personal_number = groups[13][:-1]
        personal_number_check_digit = groups[13][-1]
    
    # if the passport subtype is a filler character ("<"), set it to None
    if passport_subtype == '<':
        passport_subtype = None
    
    # return the parsed information as a dictionary
    return {
        'passport_type': passport_type,
        'passport_subtype': passport_subtype,
        'issuing_country': issuing_country,
        'surname': surname,
        'given_names': given_names,
        'passport_number': passport_number,
        'passport_number_check_digit': passport_number_check_digit,
        'nationality': nationality,
        'date_of_birth': date_of_birth,
        'date_of_birth_check_digit': date_of_birth_check_digit,
        'sex': sex,
        'expiration_date': expiration_date,
        'expiration_date_check_digit': expiration_date_check_digit,
        'personal_number': personal_number,
        'personal_number_check_digit': personal_number_check_digit
    }
```
